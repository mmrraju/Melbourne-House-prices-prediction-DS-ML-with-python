```python
'''Here I try  to take suitable tree size to avoid under fitting or overfitting. After that make a prediction model
and get new mean_absolute_error value. My target is to reduce MAE value'''
#import python library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

#load the dataset
train_data=pd.read_csv('train.csv')

#view top five row of our dataset.
train_data.head()
```

```python
#create target object
y = train_data.SalePrice

#Create feature
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = train_data[features]

#split into validation and training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=1)

#Specipy model
model = DecisionTreeRegressor(random_state=1)

#Fit the model
model.fit(X_train, y_train)

#Make validation predictions and calculate mean absolute error
valid_predictions=model.predict(X_valid)
valid_mae = mean_absolute_error(valid_predictions, y_valid)

print("Validation MAE: {:,.0f}".format(valid_mae))
```

```python
#Make a function and compare different tree size
def get_mae(max_leaf_nodes, X_train, X_valid, y_train, y_valid):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(X_train, y_train)
    valid_predictions = model.predict(X_valid)
    mae = mean_absolute_error(valid_predictions, y_valid)
    return mae

candidate_max_leaf_nodes = [5, 24, 50, 100, 250, 500]

#Making a loop to find the ideal tree size from candidate_max_leaf_nodes
best = {i:get_mae(i, X_train, X_valid, y_train, y_valid) for i in candidate_max_leaf_nodes }

for max_leaf_nodes in candidate_max_leaf_nodes:
    my_mae = get_mae(max_leaf_nodes, X_train, X_valid, y_train, y_valid)
    print("Max leaf nodes: %d \t\t Mean absolute error: %d\n"%(max_leaf_nodes, my_mae))

#Store the best value of max_leaf_nodes
best_tree_size = min(best, key=best.get)
print('Best tree size: ', best_tree_size)
```

```python
#Fit model using all Data
final_model = DecisionTreeRegressor(max_leaf_nodes = best_tree_size, random_state = 1)
final_model.fit(X_train, y_train)
final_pred = final_model.predict(X_valid)
mae = mean_absolute_error(final_pred, y_valid)
print("MAE : ", mae)
```
